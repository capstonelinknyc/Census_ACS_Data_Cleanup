```python
#create lookup table, read first two lines
aff_lookup = {}
for x in aff_txt:
    with open(aff_path+"/"+x) as myfile:
        head = [next(myfile).replace('\n','') for x in range(2)]
        aff_lookup[head[0]] = head[1].lower().replace(':', '').replace(';', '')\
              .replace('(', '').replace(')', '').replace('-','')\
              .strip().replace(' ', '_')
#remove erroreous extraction
aff_lookup.pop('', None)
print(aff_lookup)
```

```python
key = list(aff_lookup.keys())[0]
key
meta = [s for s in aff_csv if key in s][0]
print(meta)
```

```python
#create clean csv tables for backing-up
for x in aff_lookup.keys():
    meta = [s for s in aff_csv_meta if x in s][0]
    dataset = [s for s in aff_csv if x in s][0]
    print("processing metadata file : {}".format(meta))
    print("processing dataset file : {}".format(dataset))
    colz=[]         #an empty list to store the second column
    with open(aff_path+"/"+meta, 'r') as rf:
        reader = csv.reader(rf, delimiter=',')
        for row in reader:
            colz.append(row[1])
    colz_clean = [x.lower().replace(':', '').replace(';', '')\
                  .replace('(', '').replace(')', '').replace('-','')\
                  .strip().replace(' ', '_') for x in colz]
    print("printing sample cleaned columns {}".format(colz_clean[:5]))
    pd_ = pd.read_csv(aff_path+"/"+dataset,skiprows=1)
    print("table has {} rows".format(len(pd_.index)))
    pd_.columns = colz_clean
    #geography filter for "NYC"
    pd_= pd_[pd_['geography'].str.contains("NYC")]
    pd_['PUMA'] = pd_.id2.astype(str).str[3:].astype(int)
    print("saving file to {}/{}.csv".format(output_path,aff_lookup[x]))
    pd_.to_csv(output_path+"/"+aff_lookup[x].replace('/', '_').replace('/', '_')+".csv",index=False)
    print("-----------------------------------------")
```

```python
#create base table
base_table_file = list(aff_lookup.values())[0]
pd_ = pd.read_csv(output_path+"/"+base_table_file.replace('/', '_').replace('/', '_')+".csv")
base_table = pd_[['id','id2','geography']]

for x in aff_lookup.keys():
    print("processing dataset : {}".format(aff_lookup[x].replace('/', '_').replace('/', '_')))
    pd_ = pd.read_csv(output_path+"/"+aff_lookup[x].replace('/', '_').replace('/', '_')+".csv")
    drop_cols=[x for x in pd_.columns.get_values() if "margin_of_error" in x]
    #convert '' to null
    pd_.replace('', np.nan,inplace=True)
    #convert (X) to null
    pd_.replace('(X)', np.nan,inplace=True)
    #checks for any NaN values and returns column name
    drop_cols.extend(pd_.columns[pd_.isnull().any()].tolist())
    pd_=pd_.drop(drop_cols,axis=1)
    cols_to_use = pd_.columns.difference(base_table.columns)
    base_table = base_table.merge(pd_[cols_to_use],left_index=True, right_index=True, how='outer')
    print("-----------------------------------------")
```
